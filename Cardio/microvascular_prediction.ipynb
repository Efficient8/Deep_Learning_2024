```python
import pandas as pd
import numpy as np

# Assuming your dataset is stored in a DataFrame named 'df1'
original_df = df1.copy()

# Function to add noise to numeric columns
def add_noise(value, noise_level=0.05):
    if isinstance(value, (int, float)):  # Apply only to numeric values
        if value == 0 or value == 1:  # Preserve binary values
            return value
        return value * (1 + np.random.uniform(-noise_level, noise_level))
    return value  # Return non-numeric values unchanged

# Augment the dataset by duplicating rows with random noise
augmented_df = original_df.copy()
augmentation_factor = 11  # Define the number of times to duplicate each row

# Loop through each row in the original dataset and duplicate it with noise
for _ in range(augmentation_factor):
    noisy_data = original_df.copy()
    noisy_data[original_df.select_dtypes(include=['number']).columns] = \
        noisy_data[original_df.select_dtypes(include=['number']).columns].applymap(add_noise)
    augmented_df = pd.concat([augmented_df, noisy_data], ignore_index=True)

# Display the augmented dataset
augmented_df
```

```python
sex={'M':1,'F':0}
augmented_df['SEX']=augmented_df['SEX'].map(sex)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Define numerical features (including SocioeconomicStatus)
numerical_columns = [
    "AGE", "DURATION OF DM", "BMI","WHR", "FBS", "PPBS", "HbA1C",
    "TC", "TG","LDL","HDL","VLDL", "B UREA", "S CREATININE"
]

# Apply Min-Max Scaling to all numerical features
scaler = MinMaxScaler()
df2[numerical_columns] = scaler.fit_transform(df2[numerical_columns])

# Ensure target variable remains unchanged
df2["RETINOPATHY"] = df2["RETINOPATHY"].astype(int)
df2["NEPHROPATHY"] = df2["NEPHROPATHY"].astype(int)
df2["NEUROPATHY"] = df2["NEUROPATHY"].astype(int)
df2["CAD"] = df2["CAD"].astype(int)
df2["CVA"] = df2["CVA"].astype(int)
df2["PERIPHERAL ARTERIAL DISEASE"] = df2["PERIPHERAL ARTERIAL DISEASE"].astype(int)
# Display the first few rows
df2.head()
```

```python
x = df2.iloc[:, :-6]
y = df2.iloc[:, -6:-3]
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# Dictionary to store models and accuracies
models = {}
train_accuracies = {}
test_accuracies = {}

# Train separate XGBoost models for each target column
for column in y_train.columns:
    print(f"Training XGBoost model for {column}...")

    model = XGBClassifier(n_estimators=200, max_depth=6, learning_rate=0.01, objective="binary:logistic")
    model.fit(x_train, y_train[column])  # Train on a single target column

    # Predictions
    y_train_pred = model.predict(x_train)  # Training predictions
    y_test_pred = model.predict(x_test)  # Testing predictions

    # Calculate accuracies
    train_acc = accuracy_score(y_train[column], y_train_pred)
    test_acc = accuracy_score(y_test[column], y_test_pred)

    print(f"Training Accuracy for {column}: {train_acc:.4f}")
    print(f"Testing Accuracy for {column}: {test_acc:.4f}\n")

    # Store model and accuracies
    models[column] = model
    train_accuracies[column] = train_acc
    test_accuracies[column] = test_acc

# Print overall accuracy results
print("XGBoost Multi-Label Classification Training Accuracies:", train_accuracies)
print("XGBoost Multi-Label Classification Testing Accuracies:", test_accuracies)
```
